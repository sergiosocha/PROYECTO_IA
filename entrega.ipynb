```python
versus_path = "versus"

juegos = []

for filename in os.listdir(versus_path):
    if filename.endswith(".json"):
        with open(os.path.join(versus_path, filename)) as f:
            data = json.load(f)
            juegos.append((filename, data))
```

```python
def print_board(board):
    print(np.array(board))
```

```python
def get_final_state(game_steps):
    final_board, last_col = game_steps[-1]
    return final_board, last_col
```

```python
def count_moves(game_steps):
    return len(game_steps)
```

```python
def replay_game(game_steps):
    for i, (board, col) in enumerate(game_steps):
        print(f"\n=== Movimiento {i+1} (columna {col}) ===")
        print_board(board)
```

```python
def get_last_player(game_steps):
    return 1 if len(game_steps) % 2 == 1 else -1
```

```python
def show_game_info(match, game_index=0):
    filename, data = match

    print("\n======================================")
    print("Match:", filename)
    print("Player A:", data["player_a"])
    print("Player B:", data["player_b"])
    print("A wins:", data["player_a_wins"])
    print("B wins:", data["player_b_wins"])
    print("Draws:", data["draws"])
    print("Cantidad de juegos:", len(data["games"]))

    game_steps = data["games"][game_index]

    print(f"\n====== Juego {game_index + 1} ======")

    final_board, last_col = get_final_state(game_steps)

    print("\nEstado final:")
    print_board(final_board)

    print("\nÚltima columna jugada:", last_col)
    print("Total movimientos:", count_moves(game_steps))
    print("Último jugador:", get_last_player(game_steps))
```

```python
matches_F = []

for match in juegos:
    filename, data = match
    if "Group F" in data["player_a"] or "Group F" in data["player_b"]:
        matches_F.append(match)

print("Matches del Grupo F encontrados:", len(matches_F))
```

```python
for match in matches_F:
    show_game_info(match)
```
